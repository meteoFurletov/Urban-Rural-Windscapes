```python
# Loading data
df = pd.read_csv('data/ten_years/22891.csv', skiprows= 6, encoding='UTF-8', on_bad_lines='skip', sep=';', index_col=False)

# Replacing wind directions with numeric representation
df.DD.replace(wind_directions, inplace = True)

# Rename time column, change format to datetime and sort by time
print(df.columns[0])
df.rename({f'{df.columns[0]}':'time'}, axis=1, inplace=True)
df['time'] = pd.to_datetime(df['time'], dayfirst=True)
df.sort_values('time', inplace=True)
df.reset_index(drop=True, inplace=True)

# get only wind data and time, rename columns
df = df[['time', 'DD', 'Ff']]
df.rename({'DD':'wd', 'Ff':'ws'}, axis=1, inplace=True)

# create variable for df and get info about nan values
df_22891 = df
df.isna().sum()
```
